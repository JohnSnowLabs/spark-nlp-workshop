# [Thai Sentiment Classification](https://nlp.johnsnowlabs.com/2021/01/14/sentiment_jager_use_th.html)

# [Arabic Named Entity Recognition (NER)](https://nlp.johnsnowlabs.com/2020/12/05/aner_cc_300d_ar.html)
![ar](https://www.worldometers.info/img/flags/small/tn_sa-flag.gif)

```
# Arabic for : 'In 1918, the forces of the Arab Revolt liberated Damascus with the help of the British'  
nlu.load('ar.ner').predict('في عام 1918 حررت قوات الثورة العربية دمشق بمساعدة من الإنكليز',output_level='chunk')[['entities_confidence','ner_confidence','entities']]
```
